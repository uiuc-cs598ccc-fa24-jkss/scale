## Installing

```python
%%bash
set -e

cd ../../test-clusters/cash-flow
./k8s_manage.sh stop
./setup.sh
pwd
python -m venv cli
source cli/bin/activate  && pip install -r ../../test-clusters/cash-flow/cli/requirements.txt && deactivate
source cli/bin/activate  && pip install -r ../../test-clusters/cash-flow/backend/requirements.txt && deactivate
source cli/bin/activate  && pip install -e ../../test-clusters/cash-flow/cli && deactivate
./k8s_manage.sh start
cd ../../charts

helm install scale ./scale-full -n cash-flow --set sampler.trainSize=150 --set modeler.targetNamespace="cash-flow" --set modeler.configDir=cash-flow --set modeler.image.version=v1.0.4-test.1
kubectl -n cash-flow patch svc sampler -p '{"spec": {"type": "NodePort"}}'
kubectl -n cash-flow patch svc tempo -p '{"spec": {"type": "NodePort"}}'
```
